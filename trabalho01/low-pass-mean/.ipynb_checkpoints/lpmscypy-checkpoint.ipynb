```python
import cv2
import scipy.ndimage
import matplotlib.pyplot as plt

# Opening the image and converting it to grayscale.
a = cv2.imread('figs//lena.png')

# Converting the image to grayscale.
a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)

# Performing the median filter.
b = scipy.ndimage.filters.median_filter(a, size=3)

# Saving b as median_output.png in Figures folder
#cv2.imwrite('output/median_output.png', b)
```

```python
# We import matplotlib.pyplot to display an image in grayscale.
# If gray is not supplied the image will be displayed in color.
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(a, 'gray')
ax.set_title('Original')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(b, 'gray')

ax.set_title('Filtro da MÃ©dia')
```
